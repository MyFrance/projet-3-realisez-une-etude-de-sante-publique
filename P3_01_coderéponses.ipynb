```python
# Nous allons unir les DataFrames animaux et vegetaux car ils contiennent les mêmes colonnes. Cette union va nous créer 
# un troisième DataFrame animaux_vegetaux qui contiendra toutes les lignes des deux DataFrames.

animaux_vegetaux = pd.concat([animaux_df, vegetaux_df])
animaux_vegetaux.head()
```

```python
# On fait un tableau croisé dynamique  du DataFrame "animaux_vegetaux" pour que les éléments de la combinaison 
# zone et Produit soient affichés par colonne.
# Pour faire ce tableau croisé dynamique, on utilise la fonction "pivot_table"
```

```python
animaux_vegetaux_pivot = animaux_vegetaux.pivot_table("valeur", index = ["zone","code_zone","produit", 
                                                                         "code_produit", "origine"], 
                                                                columns = "element", fill_value = 0)
# Nous faisons par la même occasion un reset_index pour nous permettre de faire les calculs avec les élements de l'index

animaux_vegetaux_pivot = animaux_vegetaux_pivot.reset_index()
animaux_vegetaux_pivot
```

```python
# Avant de continuer dans nos calculs, nous allons en premier lieu renommer les colonnes du dataframe animaux_vegetaux_pivot
animaux_vegetaux_pivot_df = animaux_vegetaux_pivot.rename (columns = {"Aliments pour animaux": "Aliments_animaux",
                                                          "Autres utilisations (non alimentaire)": 
                                                          "Autres_utilisations",
                                                          "Disponibilité alimentaire (Kcal/personne/jour)": 
                                                          "Disp_alimentaire(Kcal/personne/jour)",
                                                          "Disponibilité alimentaire en quantité (kg/personne/an)":
                                                          "Dispo_alimentaire(kg/personne/an)",
                                                          "Disponibilité de matière grasse en quantité (g/personne/jour)": 
                                                          "Disp_matière_grasse(g/personne/jour)",
                                                          "Disponibilité de protéines en quantité (g/personne/jour)":
                                                          "Disp_proteines(g/personne/jour)",
                                                          "Exportations - Quantité": "Exportations",
                                                          "Importations - Quantité": "Importations",
                                                          "Variation de stock": "Variation_stock", 
                                                          "Disponibilité intérieure": "Disp_interieure",})

```

Equation à 3 termes de type (a_1 + a2 + [...] = b_1 + b_2 + [...] = c_1 + c_2 + [...]) ) faisant intervenir chacune des 11 quantités: 


Disponibilité intérieure = production + importation + variation stocks - exportation = Disp_calc1


Disponibilité intérieure = Nourriture + aliment pour animaux + traitement + semence + perte + autre utilisation (non alimentaire) = Disp_calc2

```python
animaux_vegetaux_pivot_df["Disp_interieure1"] = animaux_vegetaux_pivot_df["Production"] \
                                              + animaux_vegetaux_pivot_df["Importations"] \
                                              + animaux_vegetaux_pivot_df["Variation_stock"] \
                                              - animaux_vegetaux_pivot_df ["Exportations"]

animaux_vegetaux_pivot_df["Disp_interieure2"] = animaux_vegetaux_pivot_df ["Nourriture"] \
                                              + animaux_vegetaux_pivot_df["Aliments_animaux"] \
                                              + animaux_vegetaux_pivot_df["Traitement"] \
                                              + animaux_vegetaux_pivot_df ["Semences"] \
                                              + animaux_vegetaux_pivot_df["Pertes"] \
                                              + animaux_vegetaux_pivot_df["Autres_utilisations"]

animaux_vegetaux_pivot_df.head()
```

# Nous allons comparer les Disp_interieure1 et Disp_interieure1 avec la disponibilité intérieure animaux et vegetaux

```python
animaux_vegetaux_pivot_df["Disp_OK_1"]=animaux_vegetaux_pivot_df["Disp_interieure2"]/animaux_vegetaux_pivot_df["Disp_interieure"]
animaux_vegetaux_pivot_df["Disp_OK_2"]=animaux_vegetaux_pivot_df["Disp_interieure2"]/animaux_vegetaux_pivot_df["Disp_interieure"]
animaux_vegetaux_pivot_df.head()
```

Pour Disp_calc1 et Disp_calc2 nous allons calculer l'erreur pour 5% soit:
0.95<= Disp_calc1 <= 1.05 et 0.95<= Disp_calc2 <= 1.05
L'erreur sera pour les valeurs strictement en dessous de 0.95 et strictement au dessus de 1.05

```python
pd.concat([animaux_vegetaux_pivot_df[animaux_vegetaux_pivot_df["Disp_OK_1"] < 0.95],
         animaux_vegetaux_pivot_df[animaux_vegetaux_pivot_df["Disp_OK_1"] > 1.05 ]])

pd.concat([animaux_vegetaux_pivot_df[animaux_vegetaux_pivot_df["Disp_OK_2"] < 0.95],
         animaux_vegetaux_pivot_df[animaux_vegetaux_pivot_df["Disp_OK_2"] > 1.05 ]])

```

Donc on remarque que sur 15702 lignes  on a 756 lignes  qui ne respectent pas la formule à (756/15702 *100) = 0.53 %

# Question 3 :

Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

Vous ferez cela à partir de ces informations :

Population de chaque pays ;
Disponibilité alimentaire donnée pour chaque produit et pour chaque pays en kcal/personne/jour.

Disponibilité alimentaire en protéines donnée pour chaque produit et pour chaque pays en g/personne/jour.

Pour resoudre cet exercice, nous allons faire une jointure du dataframe animaux_vegetaux_pivot_df et filtered_Population_df.

_Le dataframe animaux_vegetaux_pivot_df est issue:
        1. de la concaténation du dataframe animaux et vegetaux.
        
        2. Et en second du tableau croisé dynamique "animaux_vegetaux" obtenu en faisant un pivot_table du dataframe "animaux_vegetaux" et de la colonne "element" du même dataframe "animaux_vegetaux
        
        3. Nous avons également filtré les données en enlevant tous les agrégats dont la Chine

_ Le dataframe filtered_Population_df est quand à lui obtenu en ne filtrant que les "données officielles" du dataframe population, en appliquant un mask à la colonne "Description_Symbole"= "données officielles" ce qui  nous a permis  de supprimer toutes les données "Agrégat, peut inclure des données officielles, semi-officielles, estimées ou calculées".Donc la chine 

Pour ce faire la condition de jointure sera pour le dataframe animaux_vegetaux "code_zone" et pour le dataframe filtered_Population_df "code_zone.

La colonne commune qui effectue une correspondance entre les deux dataframes est donc la colonne "code_zone"

Ici la condition de jointure dit que c'est la colonne Code_zone de animaux_vegetaux qui reférence la colonne produit de dataframe population

```python
# Jointure du dataframe animaux_vegetaux et de filtered_Population_df 

# animaux_vegetaux_pop = Bilan alimentaire

animaux_vegetaux_pop = pd.merge(animaux_vegetaux_pivot_df, filtered_population_df, left_on =['zone','code_zone'],
                                                                                  right_on =['zone','code_zone'] )
#left_on ='code_zone', right_on ='code_zone' 

animaux_vegetaux_pop.head()

```

Calcul de la disponibilité alimentaire en kcal pour chaque pays et pour chaque produit  

Disponibilité alimentaire( Kcal)=  Disponibilité alimentaire (Kcal/personne/jour) * 365 * Population_totale 

```python
animaux_vegetaux_pop["Disp_alim_AniVegPop_kcal"] = animaux_vegetaux_pop["Disp_alimentaire(Kcal/personne/jour)"]*\
                                                  365* animaux_vegetaux_pop["Population_totale"]
animaux_vegetaux_pop.head()
```

Calcul de la disponibilité alimentaire en protéine_kg pour chaque pays et pour chaque produit 

Disponibilité alimentaire( Kg)=  Disponibilité de protéines en quantité (g/personne/jour)/1000 * 365 * Population_totale)

```python
# Disponibilité alimentaire en protéine kg pour chaque pays et pour chaque produit
animaux_vegetaux_pop["Disp_prot_AniVegPop_kg"] = animaux_vegetaux_pop["Disp_proteines(g/personne/jour)"]\
                                                /1000 * 365* animaux_vegetaux_pop["Population_totale"]
animaux_vegetaux_pop.head()
```
