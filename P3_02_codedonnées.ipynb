 Pour quelques uns des produits identifiés  quelles sont les "autres utilisations" possibles.


1. ALCOOLS NON COMESTIBLES utilisés dans l’industrie pharmaceutiques, industrie chimiques et l’industrie cosmétiques. 


## Dans l’industrie pharmaceutique et cosmétique:

on l’utilise comme solvant sous forme d’éthanol.

Il est miscible dans l'eau et dans la plupart des autres solvants usuels. 

Il est stable, incolore, volatile et inflammable. Il peut agir en tant que désinfectant. 

## Dans l’industrie chimique:

on l’utilise soit sous forme d’éthanol mais aussi sous forme d’alcool dénaturé dont le méthanol,

le phtalate de diéthyle, le thiophène etc. 

Avec ces éléments on peut faire différentes réactions chimiques pour l’obtention d’autres composants chimiques.


On peut aussi utilisé dans nos maisons comme alcool à bruler ou alcool ménager


2. L’HUILE DE PALME 


L’huile de palme est beaucoup utilisé dans l’agroalimentaire mais aussi en cosmétique.

## En agroalimentaire

On le retrouve dans les plats cuisinés, les friandises, les pâtes à tartiner.

## En cosmétique 

on le retrouve sous ces différents noms: 

 - PALM- désignant l’acide palmitique ou palmate
 
 - STEAR- pour l’acide stéarique ou stéarate
 
 - LAURYL- pour le Lauryl polyglucose par exemple

On le retrouve dans les gels douches, la mousse à raser, les fonds de teint, les mascaras et crayons à yeux.
